# Some helpful documentation about pandas dataframes

https://medium.com/dunder-data/selecting-subsets-of-data-in-pandas-6fcd0170be9c

`df[ ]`   It can select subsets of columns.


`df.loc[ ]` It can select subsets of rows or columns. It can also simultaneously select subsets of rows and columns. __Most importantly, it only selects data by the LABEL of the rows and columns.__      

`df.iloc[ ]`  Very similar to .loc __but only uses integer locations__ to make its selections. The word .iloc itself stands for integer location so that should help with remember what it does.

Selecting multiple specific columns: `college_data[['Accept', 'PhD']]`

Selecting rows and columns simultaneously : `df.loc[row_selection, column_selection]`  

Selecting specific rows and columns by index with iloc: `df.iloc[[2,3], [0, 4]]`


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 8a. Use theread.csv()function to read the data intoR. Calltheloaded datacollege.
csv_path = r'C:\Users\The_Iron_Maiden\Documents\Python Scripts\College.csv'

college_data = pd.read_csv(csv_path)
```

```python
college_data.head(3)
```

```python
# 8b. Use university name as index
college_data = college_data.set_index(['Unnamed: 0'], 'name')
college_data.head(3)
```

```python
# 8c.i : use thesummary()function to produce a numerical summaryof the variables in the data set.

college_data.describe()
# doesn't show categorical data
```

```python
# 8c.ii. Produce a scatterplot matrix of the first ten columns or variables of the data

axes = pd.plotting.scatter_matrix(college_data.iloc[:,:11], figsize=(10,10))
# ignores categorical data
```

```python
# 8c.iii : Use theplot()function to produce side-by-side boxplots ofOutstateversusPrivate

fig, axs = plt.subplots(1, 2, figsize = (9,4))
axs[0].boxplot(college_data['Outstate'])
axs[0].set_title('Outstate Tuition')
axs[0].set_ylabel('Dollars')

axs[1].hist(college_data['Private']);
axs[1].set_title('Private Institute')
axs[1].set_ylabel('Count')
```

```python
# 8c.iv : create a new qualitative variable, calledElite,bybinningtheTop10percvariable

# make a vector of No's and add as column in dataframe
college_data['elite'] = 'No'
```

```python
college_data['Top10perc']>50
```

```python
# elite: if 50% of student body were top 10% in high school
"""8c.iv : We are going to divide universitiesinto two groups based on whether or not the proportionof students 
 coming from the top 10 % of their high schoolclasses exceeds 50 %
"""

# for the elite column, find rows where 'Top10perc' above 50, turn to YES
elite_thresh = 50 # in percentage
college_data['elite'].loc[college_data['Top10perc'] > elite_thresh] = 'Yes' # loc seems to violate intro statements (not really referencing column name)
college_data[['Top10perc', 'elite']].head()
```

```python
# 8c.iv : plot results

print(college_data['elite'].describe())

fig, axs = plt.subplots(1, 1, figsize = (4,4))
axs.hist(college_data['elite'])
axs.set_title('Elite')
```

# Boston data


    CRIM - per capita crime rate by town
    ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
    INDUS - proportion of non-retail business acres per town.
    CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
    NOX - nitric oxides concentration (parts per 10 million)
    RM - average number of rooms per dwelling
    AGE - proportion of owner-occupied units built prior to 1940
    DIS - weighted distances to five Boston employment centres
    RAD - index of accessibility to radial highways
    TAX - full-value property-tax rate per $10,000
    PTRATIO - pupil-teacher ratio by town
    B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    LSTAT - % lower status of the population
    MEDV - Median value of owner-occupied homes in $1000's

```python
"""
10d. Do any of the suburbs of Boston appear to have particularly high crime rates? 
Tax rates? Pupil-teacher ratios? Comment on the range of each predictor.
"""
n_top = 3
features = ['crim', 'tax', 'ptratio']

for feature in features:
    print(feature)
    print(boston_data.nlargest(n_top, feature))
    print(' ')
```
